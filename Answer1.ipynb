```python
# ================================
# UNIVERSITY STUDENT PERFORMANCE ANALYSIS
# Using: Pandas + NumPy (Built-in Dataset)
# ================================

import pandas as pd
import numpy as np

# -------------------------------
# 1. DATASET CREATION
# -------------------------------

np.random.seed(42)
n_students = 200

df = pd.DataFrame({
    "Student_ID": np.arange(1, n_students + 1),
    "Study_Hours_per_Week": np.random.normal(15, 5, n_students).round(1),
    "Attendance_Percentage": np.random.randint(50, 101, n_students),
    "Assignment_Score": np.random.randint(40, 101, n_students),
    "Midterm_Score": np.random.randint(35, 101, n_students)
})

# Ensure no negative study hours
df["Study_Hours_per_Week"] = df["Study_Hours_per_Week"].clip(lower=0)

# -------------------------------
# 2. FINAL SCORE CALCULATION
# -------------------------------

df["Final_Score"] = (
    0.3 * df["Assignment_Score"] +
    0.3 * df["Midterm_Score"] +
    0.4 * (df["Study_Hours_per_Week"] * 2)
).round(1)

# -------------------------------
# 3. GRADE CLASSIFICATION
# -------------------------------

df["Grade"] = np.where(df["Final_Score"] >= 85, "A",
                np.where(df["Final_Score"] >= 70, "B",
                np.where(df["Final_Score"] >= 55, "C", "Fail")))

# -------------------------------
# 4. AT-RISK IDENTIFICATION
# -------------------------------

df["At_Risk"] = np.where(
    (df["Attendance_Percentage"] < 60) &
    (df["Study_Hours_per_Week"] < 10) &
    (df["Final_Score"] < 55),
    "Yes",
    "No"
)

# -------------------------------
# 5. GROUP CATEGORIZATION
# -------------------------------

df["Attendance_Group"] = np.where(
    df["Attendance_Percentage"] >= 80, "High", "Low"
)

df["Study_Category"] = pd.cut(
    df["Study_Hours_per_Week"],
    bins=[0, 10, 20, 40],
    labels=["Low", "Medium", "High"]
)

# -------------------------------
# 6. ANALYSIS SECTION
# -------------------------------

print("===== BASIC STATISTICS =====")
print("Class Average Final Score:", df["Final_Score"].mean())
print("\nGrade Distribution:\n", df["Grade"].value_counts())

print("\n===== CORRELATION MATRIX =====")
print(df.corr(numeric_only=True))

print("\n===== PERFORMANCE BY ATTENDANCE GROUP =====")
print(df.groupby("Attendance_Group")["Final_Score"].mean())

print("\n===== PERFORMANCE BY STUDY CATEGORY =====")
print(df.groupby("Study_Category")["Final_Score"].mean())

print("\n===== TOP 5 PERFORMERS =====")
print(df.sort_values("Final_Score", ascending=False).head())

print("\n===== BOTTOM 5 PERFORMERS =====")
print(df.sort_values("Final_Score").head())

print("\n===== AT RISK STUDENTS =====")
print(df[df["At_Risk"] == "Yes"].head())

# -------------------------------
# 7. SIMPLE RULE-BASED PREDICTION
# -------------------------------

df["Prediction"] = np.where(
    (df["Study_Hours_per_Week"] > 15) &
    (df["Attendance_Percentage"] > 75),
    "Pass",
    "Risk"
)

accuracy = np.mean(
    ((df["Prediction"] == "Pass") & (df["Final_Score"] >= 55)) |
    ((df["Prediction"] == "Risk") & (df["Final_Score"] < 55))
)

print("\n===== SIMPLE MODEL ACCURACY =====")
print("Accuracy:", round(accuracy * 100, 2), "%")

# -------------------------------
# END OF PROJECT
# -------------------------------
```
