```python
import pandas as pd

df = pd.read_csv("Dataset  - Sheet1.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
top_performers = df[df["Score"] > 90]
print(top_performers)
sales_data = df[df["Department"] == "Sales"]
print(sales_data)
df_sorted = df.sort_values(by="Score", ascending=False)
print(df_sorted)
df_dept_sorted = df.sort_values(by=["Department", "Score"], ascending=[True, False])
print(df_dept_sorted)
avg_dept_score = df.groupby("Department")["Score"].mean()
print(avg_dept_score)
max_quarter_score = df.groupby("Quarter")["Score"].max()
print(max_quarter_score)
```
